```python
import numpy as np
import pandas as pd
import time
import datetime

np.random.seed(12)

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
matplotlib.rcParams['axes.unicode_minus']=False
sns.set_style("darkgrid",{"font.sans-serif":['simhei', 'Arial']})

pd.options.mode.chained_assignment = None  # default='warn'
pd.options.display.max_columns = 999
```

```python
data = pd.read_csv('data/data_set_v1.csv')
```

```python
data = data.drop(['scopus','doi','title','source','link','affiliation','pages','volume','given-name','sur_name'],axis=1)
data.drop_duplicates()
data.info()
```

```python
data = data.dropna(how='any')
data.info()
```

Mendeley阅读数realer_count, 作者被引用总数citedby_count + h因子h_index + 总文章数doc_total + 文章应用总数cit_by_doc + 总合作作者数co_authors_total

```python
sns.pairplot(data, x_vars=['citedby_count','doc_total','cit_by_doc','co_authors_total'], y_vars='reader_count', size=7, aspect=0.8)
plt.show()
```

```python
g = sns.lmplot(x='citedby_count', y='reader_count',col='year',data=data, col_wrap=2, size=3)

g = g.set(xlim=(0, 800), ylim=(0, 750),xticks=[200, 400, 800], yticks=[250, 500, 750]).fig.subplots_adjust(wspace=.02)
```

```python
data0 = data[['reader_count','citedby_count','doc_total','cit_by_doc','co_authors_total']]

corr = data0.corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
X = data[['citedby_count','doc_total','cit_by_doc','co_authors_total']]
y = data[['reader_count']]

from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, matthews_corrcoef

X_train,X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)  

from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
model=linreg.fit(X_train, y_train)

y_pred = model.predict(X_test)
print('自变量系数为: \n', model.coef_)
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
print('Variance score: %.3f' % r2_score(y_test, y_pred))
```

```python
X = data[['citedby_count']]
y = data[['reader_count']]

X_train,X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

model=linreg.fit(X_train, y_train)

y_pred = model.predict(X_test)
print('自变量系数为: \n', model.coef_)
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
print('Variance score: %.3f' % r2_score(y_test, y_pred))
```
